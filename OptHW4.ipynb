### Probelm 3

In optimization with inequality constraints, binding means an inequality constraint is effecting the solution, which essentially becomes a equality constraint. Not binding means the inequality constraint is not effecting. 

For example, in the following problem
\begin{gather}
    \min_{x} x^2 \\
    \text{s.t.} \nonumber\\
    x \leq 2
\end{gather}
the inequality constraint $x \leq 2$ is not binding because the optimal solution is $x^* = 0$.

While in this problem
\begin{gather}
    \min_{x} x^2 \\
    \text{s.t.} \nonumber\\
    x \geq 1
\end{gather}
the inequality constraint is  binding because the optimal solution is $x^* = 1$.
